## Загрузка файла CSV с ID в качестве индекса

```python
import pandas as pd

df = pd.read_csv('data/auto.csv', index_col='ID')

df.head()
```

## Подсчет наблюдений и удаление дубликатов

```python
print(f'Количество непустых наблюдейний в каждом столбце:\n{df.count()}')

df = df.drop_duplicates(subset=['CarNumber', 'Make_n_model', 'Fines'], keep='last')
print(f'\nКоличество непустых наблюдений в каждом столбце после удаления дубликатов:\n{df.count()}')
```

## Работа с пропущенными значениями - удаление столбцов со многими NA

```python
print(f'Отсутствующие значения до чистки:\n{df.isna().sum()}')
df = df.dropna(axis=1, thresh=len(df)-500)
print(f'\nОтсутствующие значения после чистки:\n{df.isna().sum()}')
```

## Заполнение пропущенных значений в столбцах Refund и Fines

```python
df['Refund'] = df['Refund'].ffill()
print(f'После препроцессинга Refund:\n{df.isna().sum()}')

df['Fines'] = df['Fines'].fillna(df['Fines'].mean())
print(f'\nПосле препроцессинга Fines:\n{df.isna().sum()}')
```

## Разделение Make_n_model на столбцы Make и Model

```python
df['Make'] = df['Make_n_model'].apply(lambda x: x.split()[0] if pd.notna(x) else None)
df['Model'] = df['Make_n_model'].apply(lambda x: ' '.join(x.split()[1:]) if pd.notna(x) and len(x.split()) > 1 else None)
df = df.drop(columns=['Make_n_model'])
df.head(5)
```

## Сохранение в файл JSON

```python
df.to_json('data/auto.json', orient='records', indent=4)
```

## Загрузка JSON и установка индекса

```python
import pandas as pd

df = pd.read_json('data/auto.json', orient='records')
df = df.set_index('CarNumber')
df.head(5)
```

## Выборка - штрафы больше 2100

```python
f_over2100 = df[df['Fines'] > 2100]
f_over2100
```

## Выборка - штрафы > 2100 и возврат = 2

```python
f_over2100[f_over2100['Refund'] == 2]
```

## Выборка - модели Фокус и Королла

```python
models=['Focus','Corolla']
df.loc[df['Model'].isin(models)]
```

## Выборка - конкретные номера автомобилей

```python
cnum_list = ['Y7689C197RUS', '92928M178RUS', '7788KT197RUS', 'H115YO163RUS', 'X758HY197RUS']
df[df.index.isin(cnum_list)]
```

## Агрегация - медианные штрафы по маркам

```python
df.groupby('Make')['Fines'].median()
```

## Агрегация - медианные штрафы по маркам и моделям

```python
df.groupby(['Make','Model'])['Fines'].median()
```

## Агрегация - подсчет количества штрафов по маркам и моделям

```python
df.groupby(['Make','Model']).agg('Fines').count()
```

## Агрегация - минимальные и максимальные штрафы по маркам и моделям

```python
df.groupby(['Make','Model'])['Fines'].agg(['min','max'])
```

## Агрегация - стандартное отклонение штрафов по маркам и моделям

```python
df.groupby(['Make','Model'])['Fines'].std()
```

## Агрегация номеров автомобилей по количеству штрафов

```python
carnum_sort = df.groupby(df.index)['Fines'].count().sort_values(ascending=False)
carnum_sort
```

## Топ-1 по количеству штрафов

```python
df.loc[carnum_sort.index[0]]
```

## Номера автомобилей, сгруппированные по сумме штрафов

```python
carnum_f_sort = df.groupby(df.index)['Fines'].sum().sort_values(ascending=False)
carnum_f_sort
```

## Выбор топ-1 номера автомобиля по сумме штрафов

```python
df.loc[carnum_f_sort.index[0]]
```

## Проверка номеров автомобилей, связанных с разными моделями

```python
diff_models = df.groupby(df.index)['Model'].nunique(dropna=False)
df[df.index.isin(diff_models[diff_models > 1].index)]
```

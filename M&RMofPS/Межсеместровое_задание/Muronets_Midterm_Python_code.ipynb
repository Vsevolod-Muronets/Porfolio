```python
### Problem 1
import numpy as np
import pandas as pd
import statsmodels.api as sm
from plotnine import *
np.random.seed(12345)
```

```python
## III.
Educ = []
Turn = []
for Education in range(1000):
  epsilon = np.random.normal(0.0, np.sqrt(0.01))
  Education = np.random.uniform(0.0, 15.0)
  Turnout = 0.4 + 0.01 * Education + epsilon
  Educ.append(Education)
  Turn.append(Turnout)

data_vars = pd.DataFrame({'Constant':[1]*len(Educ),
                          'IV': Educ})

model = sm.OLS(Turn, data_vars)

results = model.fit()
print(results.summary())
```

```python
data_vars['Fitted Values'] = np.dot(data_vars,results.params)

data_vars['DV'] = Turn

plot = (ggplot(data_vars, aes(x = 'IV'))+
        geom_point(aes(y = 'DV'))+
        geom_line(aes(y = 'Fitted Values'))+
        theme_bw()+
        ylab('Turnout')+
        xlab('Level of education'))
plot
```

```python
## IV.

Educ1 = []
Turn1 = []
for Education in range(1000):
  epsilon = np.random.normal(0.0, np.sqrt(0.2))
  Education = np.random.uniform(0.0, 15.0)
  Turnout = 0.4 + 0.01 * Education + epsilon
  Educ1.append(Education)
  Turn1.append(Turnout)

print(Turn1)
```

```python
print('With such a big variance of epsilon the epsilon itself becomes quite large in many observations. We can see that this results in a quite significance of the variance in Turnout (our independent variable) as well. This shows that epsilon, representing unobserved effects, actually affects the dependent variable quite significantly, and therefore in becomes hard to say that there is indeed a causal relationship between the independent variable and the dependent one, this being represented in a vary small value of R-squared in the results if we run the sm.OLS fucntion as below (in the task we are asked not to do the regression, so this is just for an illustration):')
```

```python
data_vars1 = pd.DataFrame({'Constant':[1]*len(Educ1),
                          'IV': Educ1})

model1 = sm.OLS(Turn1, data_vars1)

results1 = model1.fit()
print(results1.summary())
```

```python
### Problem 8

import numpy as np
import pandas as pd
import statsmodels.api as sm
np.random.seed(12345)
```

```python
## I
xs10 = []
ys10 = []

for x in range(10):
  epsilon = np.random.normal(0.0, np.sqrt(3.0))
  y = 1.4 + 0.4 * x + epsilon
  xs10.append(x)
  ys10.append(y)

data81_vars = pd.DataFrame({'Constant':[1]*len(xs10),
                          'Independent variable values': xs10})
model81 = sm.OLS(ys10, data81_vars)

results81 = model81.fit()
print(results81.summary())
```

```python
print('We can see that the intercept (1.2901 when the sm.OLS function was run) and the slope (0.6149 when the function was run) for this 10-observations sample are quite substantially different from those of the given population (1.4 and 0.4 respectively)')
```

```python
## II

xs1000 = []
ys1000 = []

for x in range(1000):
  epsilon = np.random.normal(0.0, np.sqrt(3.0))
  y = 1.4 + 0.4 * x + epsilon
  xs1000.append(x)
  ys1000.append(y)

data82_vars = pd.DataFrame({'Constant':[1]*len(xs1000),
                          'Independent variable values': xs1000})
model82 = sm.OLS(ys1000, data82_vars)

results82 = model82.fit()
print(results82.summary())
```

```python
print('We can see that the intercept (1.3219 when the function was run) and the slope (0.4001 when the function was run) for this 1000-observations sample are more correct than those for the 10-observations sample, though still different from those of the given population (1.4 and 0.4 respectively)')
```

```python
## III

xs10000 = []
ys10000 = []

for x in range(10000):
  epsilon = np.random.normal(0.0, np.sqrt(3.0))
  y = 1.4 + 0.4 * x + epsilon
  xs10000.append(x)
  ys10000.append(y)

data83_vars = pd.DataFrame({'Constant':[1]*len(xs10000),
                          'Independent variable values': xs10000})
model83 = sm.OLS(ys10000, data83_vars)

results83 = model83.fit()
print(results83.summary())
```

```python
print('We can see that the intercept (1.4160 when the function was run) and the slope (0.4000 when the function was run) for this 10000-observations sample are very close to those of the given population (1.4 and 0.4 respectively). This shows that the OLS (ordinary least squares) estimator works here, with the E[u] (with represented by the epsilon in the formula above) coming closer to zero with the growth of the number observations in the sample and with E[u] being equal to zero for the population')
```

```python
# Problem 4.


import numpy as np

# I.

ff1 = np.random.seed(12345)
ff1

# II.

ff2 = np.random.normal(1, 5, 10000)

# III.

s = 0

# IV.

for x in range (100):
  sample = np.random.choice(10000, size = 100, replace = False)
  ff3 = ff2[sample]
  s_mean=np.mean(ff3)
  s = s_mean + s

# V.

p_mean = np.mean(ff2)

def comparison_mean():
  s2 = s/100
  if s2>p_mean:
    calc1 = s2 - p_mean
    print("Sample mean is bigger than population mean by", calc1);
  elif s2<p_mean:
    calc2 = p_mean - s2
    print('Sample mean is lower than population mean by', calc2)
  else:
    print("Sample mean is equal to population mean");

comparison_mean()
```

```python
# VI.

ff2 = np.random.normal(1, 5, 10000)
s = 0

for x in range (100):
  sample = np.random.choice(10000, size = 100, replace = False)
  ff3 = ff2[sample]
  s_var=np.var(ff3, ddof=1)
  s = s_var + s

p_var = np.var(ff2, ddof=0)

def comparison_var():
  s2 = s/100
  if s2>p_var:
    calc1 = s2 - p_var
    print("Sample variance is bigger than population variance by", calc1);
  elif s2<p_var:
    calc2 = p_var - s2
    print('Sample variance is lower than population variance by', calc2)
  else:
    print("Sample variance is equal to population variance");

comparison_var()
```
